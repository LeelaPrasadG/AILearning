```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
(X_train,y_train), (X_test, y_test) = keras.datasets.mnist.load_data()
```

```python
X_train = X_train/255
X_test = X_test/255
```

```python
X_train_flattened = X_train.reshape(len(X_train), 28*28)
X_test_flattened = X_test.reshape(len(X_test), 28*28)
```

```python
X_test_flattened.shape
```

```python
X_train[0]
```

```python
model = keras.Sequential([
     keras.layers.Dense(10, input_shape=(784,), activation = 'sigmoid')
])

model.compile(
    optimizer='adam', 
    loss = 'sparse_categorical_crossentropy',
    metrics = ['accuracy']
)

model.fit(X_train_flattened, y_train, epochs=5)
```

```python
model.evaluate(X_test_flattened, y_test)
```

```python
plt.matshow(X_test[0])
```

```python
y_test_predicted = model.predict(X_test_flattened)
```

```python
np.argmax(y_test_predicted[0])
```

```python
y_test_predicted_labels = [np.argmax(i) for i in y_test_predicted]
```

```python
y_test_predicted_labels
```

```python
cm = tf.math.confusion_matrix(labels=y_test, predictions = y_test_predicted_labels)
cm
```

```python
import seaborn as sn
plt.figure(figsize = (10,7))
sn.heatmap(cm, annot = True, fmt = 'd')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
###Adding Another Layer to improve accuracy

model = keras.Sequential([
     keras.layers.Dense(100, input_shape=(784,), activation = 'relu'),
    keras.layers.Dense(10, activation = 'sigmoid')
])

model.compile(
    optimizer='adam', 
    loss = 'sparse_categorical_crossentropy',
    metrics = ['accuracy']
)

model.fit(X_train_flattened, y_train, epochs=5)
```

```python
model.evaluate(X_test_flattened, y_test)
```

```python
y_test_predicted = model.predict(X_test_flattened)
y_test_predicted_labels = [np.argmax(i) for i in y_test_predicted]
cm = tf.math.confusion_matrix(labels=y_test, predictions = y_test_predicted_labels)
import seaborn as sn
plt.figure(figsize = (10,7))
sn.heatmap(cm, annot = True, fmt = 'd')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
###Now calling the same came code without flattening the Array

model = keras.Sequential([
    keras.layers.Flatten(input_shape = (28,28)),
     keras.layers.Dense(100, input_shape=(784,), activation = 'relu'),
    keras.layers.Dense(10, activation = 'sigmoid')
])

model.compile(
    optimizer='adam', 
    loss = 'sparse_categorical_crossentropy',
    metrics = ['accuracy']
)

model.fit(X_train, y_train, epochs=5)
```
