```python
import pandas as pd
df = pd.read_csv("/content/sample_data/titanic.csv")
df
```

```python

df = df[['Pclass','Sex','Age','Fare','Survived']]
df
```

```python
target_df = df.Survived
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Sex_le'] = le.fit_transform(df.Sex)
preprocessed_df = df.drop(['Survived','Sex'], axis = 'columns')
```

```python
preprocessed_df
```

```python
for feature in preprocessed_df.columns:
  print('Total nulls for the column {} are {}'.format(feature,sum(preprocessed_df[feature].isna())))
```

```python
preprocessed_df['Age'] = preprocessed_df['Age'].fillna(preprocessed_df.Age.mean())
```

```python
preprocessed_df
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(preprocessed_df, target_df, test_size=0.2)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```
