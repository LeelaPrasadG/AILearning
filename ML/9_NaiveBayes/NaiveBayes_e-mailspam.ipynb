```python
import pandas as pd
from sklearn.naive_bayes import MultinomialNB
```

```python
df = pd.read_csv('/content/sample_data/spam.csv')
df
```

```python
df.groupby('Category').describe()
```

```python
df['Category'] = df['Category'].apply(lambda x: 1 if x == 'spam' else 0)
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
corpus = [
    'This is the first document.',
    'This document is the second document.',
    'And this is the third one.',
    'Is this the first document?',
]
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(corpus)
print(vectorizer.get_feature_names_out())
print(X.toarray())
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.Message, df.Category, test_size = 0.2)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
X_train_count = vectorizer.fit_transform(X_train.values)
X_train_count.toarray()
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(X_train_count, y_train)
```

```python
X_test_count = vectorizer.transform(X_test.values)
X_test_count.toarray()
predict_y_test = model.predict(X_test_count)
```

```python
predict_y_test
```

```python
y_test
```

```python
model.score(X_test_count,y_test)
```

# Usage of Pipeline to avoid calling transform and passing count to Predict function.

```python
from sklearn.pipeline import Pipeline
clf = Pipeline([
    ('vectorizer',CountVectorizer()),
    ('nb', MultinomialNB())
])
```

```python
clf.fit(X_train, y_train)
```

```python
ytest_predict = clf.predict(X_test)
```

```python
clf.score(X_test, y_test)
```
