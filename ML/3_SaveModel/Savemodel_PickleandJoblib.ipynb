```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('/content/sample_data/homeprices.csv')
df
```

```python
reg = LinearRegression()
model = reg.fit(df[['area']],df.price)
```

```python
model.predict([[4200]])
```

```python
model.coef_
```

```python
model.intercept_
```

Save Model To a File Using Python Pickle

```python
import pickle
with open('/content/sample_data/model_pickle', 'wb') as file:
  pickle.dump(model,file)
```

```python
#from sklearn.externals import joblib
import joblib
with open('/content/sample_data/model_joblib', 'wb') as file:
  joblib.dump(model,file)
```

<h4 style='color:purple'>Load Saved Model</h4>

```python
loaded_model = joblib.load('/content/sample_data/model_joblib','r+')
```
