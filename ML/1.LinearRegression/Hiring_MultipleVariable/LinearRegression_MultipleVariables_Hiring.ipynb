```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import math
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/sample_data/hiring.csv')
```

```python
df
```

```python
for feature in df.columns:
  print(df[feature].dtypes)
```

```python
categorical_features = [features for features in df.columns if df[features].dtypes == 'O']
categorical_features
```

```python
dataset = df.copy()
df.rename(columns = {'test_score(out of 10)':'test_score', 'interview_score(out of 10)':'interview_score','salary($)':'salary'}, inplace=True)
df
```

```python
median_testscore = df.test_score.median(numeric_only=True)
median_testscore
df.test_score = df.test_score.fillna(median_testscore)
df
```

```python
### Convert the categorical_features to enum
dataset = df.copy()
for feature in categorical_features:
    labels_ordered=dataset.groupby([feature])['salary'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    print(labels_ordered)
    dataset[feature]=dataset[feature].map(labels_ordered)

dataset
```

```python
##Now replace the Nan values in the categorical_features column with median


##for feature in categorical_features:
##  df[feature] = df[feature].fillna(math.floor(df[feature].median()))

###       OR    Replace NaN with 0 as the candidates has no experience      ###
df['experience'] = df['experience'].fillna(0)
```

```python
df
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['experience','test_score','interview_score']],df['salary'])
#reg.fit(df[['area','bedrooms','age']],df['price'])
```

```python
reg.predict([[2,10,10]])
```
